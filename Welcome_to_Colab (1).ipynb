```
from os import pathconf
def frequency_count(data):
    freq = {}

    for num in data:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1

    return freq

a = [10,10,10,6,7,7,6,8,3,3]
freq = frequency_count(a)
print(freq)

mode = None
highest = 0

for key in freq:
    if freq[key] > highest:
        highest = freq[key]
        mode = key
print(mode)
```

```
def reverse_number(n):
    rev = 0

    while n > 0:
        digit = n % 10
        rev = rev * 10 + digit
        n = n // 10

    return rev
print(reverse_number(123))
```

```
def is_palindrome(n):
    return n == reverse_number(n)
```

```
def tuple_to_dict(tuples_list):
    result = {}

    for pair in tuples_list:
        key = pair[0]
        value = pair[1]
        result[key] = value

    return result

print(tuple_to_dict([("a",1),("b",2),("c",3)]))
```

```
def max_key(data):
    max_k = None
    max_v = -float("inf")

    for key in data:
        if data[key] > max_v:
            max_v = data[key]
            max_k = key

    return max_k

print(max_key({"a":1,"b":2,"c":3}))
```

```
def sum_avg(*args):
    total = 0
    count = 0

    for num in args:
        total += num
        count += 1

    avg = total / count

    return total, avg

print(sum_avg(1,2,3,4,5))
```

```
def top_student(**kwargs):
    max_name = None
    max_marks = -1

    for name in kwargs:
        if kwargs[name] > max_marks:
            max_marks = kwargs[name]
            max_name = name

    return max_name

print(top_student(a=10,b=20,c=30))
```

```
class Vector:
    def __init__(self, values):
        self.values = values

    def __add__(self, other):
        result = []
        for i in range(len(self.values)):
            result.append(self.values[i] + other.values[i])
        return Vector(tuple(result))

    def __sub__(self, other):
        result = []
        for i in range(len(self.values)):
            result.append(self.values[i] - other.values[i])
        return Vector(tuple(result))

    def __mul__(self, scalar):
        result = []
        for i in range(len(self.values)):
            result.append(self.values[i] * scalar)
        return Vector(tuple(result))

    def __repr__(self):
        return f"Vector{self.values}"


v1 = Vector((1, 2, 3))
v2 = Vector((4, 5, 6))

print(v1 + v2)
print(v1 - v2)
print(v1 * 2)
```

```
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def grade(self):
        if self.marks >= 80:
            return "A"
        elif self.marks >= 60:
            return "B"
        else:
            return "C"

    def __str__(self):
        return f"Name: {self.name}, Marks: {self.marks}, Grade: {self.grade()}"

print(Student("John", 90))
```
