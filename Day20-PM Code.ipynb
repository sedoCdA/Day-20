```
import numpy as np

data = np.random.normal(50, 10, 1000)

# manual mean
total = 0
for x in data:
    total += x
mean = total / len(data)

# manual variance
var_sum = 0
for x in data:
    var_sum += (x - mean) ** 2
variance = var_sum / len(data)

std = variance ** 0.5

print(mean, variance, std)
```

```
import matplotlib.pyplot as plt

plt.hist(data, bins=30)
plt.title("Normal Distribution")
plt.show()
```

```
z_data = []

for x in data:
    z = (x - mean) / std
    z_data.append(z)

# verify
print(sum(z_data)/len(z_data))  # ≈ 0
```

```
data = [10,20,30,40,100]

# reuse mean, std from above logic

outliers = []

for x in data:
    z = (x - mean) / std
    if abs(z) > 2:
        outliers.append(x)

print("Outliers:", outliers)
```

```
import math

mu = 50  # assumed mean

z_stat = (mean - mu) / (std / (len(data) ** 0.5))

print("Z:", z_stat)

if abs(z_stat) > 1.96:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

```
count = 0
trials = 1000

for _ in range(trials):
    sample = np.random.normal(50, 10, 30)

    # compute mean
    total = 0
    for x in sample:
        total += x
    m = total / len(sample)

    # compute std
    var_sum = 0
    for x in sample:
        var_sum += (x - m) ** 2
    s = (var_sum / len(sample)) ** 0.5

    z = (m - 50) / (s / (len(sample) ** 0.5))

    if abs(z) > 1.96:
        count += 1

print("False positive rate:", count / trials)
```
